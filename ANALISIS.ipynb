```python
def desanidar(ruta, diccionario):
    """Funcion para leer json como texto y poder pasarlos a dataframe

    Args:
        ruta (str): ruta del archivo json
        diccionario (dict): diccionario con las claves que necesitamos

    Returns:
        DataFrame: retorna un dataframe desanidado listo para trabajar con pandas
    """
    df = []
    with open(ruta, encoding="utf-8-sig") as file:
        for linea in file:
            df.append(eval(linea))
    for i in df:
        for clave,valor in i.items():
            diccionario[clave].append(valor)
    diccionario = pd.DataFrame(diccionario)
    return diccionario
```

```python
def ceros(x):
    """Funcion para cambiar cualquier valor no
    numerico a 0 (Free, Free to play, etc)

    Args:
        x (str): Si es un numero lo convierte sino
        lo pasa a 0

    Returns:
        El numero convertido o 0
    """
    try:
        return float(x)
    except:
        return 0

games['price'] = games['price'].apply(ceros)
games['price']
```

```python
#CREACION DE LA COLUMNA sentiment_analysis
from textblob import TextBlob

def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity < 0:
        return 0  
    elif analysis.sentiment.polarity == 0:
        return 1
    else:
        return 2

reviews_reviews['sentiment_analysis'] = reviews_reviews['review'].apply(analyze_sentiment)
reviews_reviews['sentiment_analysis'].fillna(1, inplace=True)
```

```python
# Debe devolver cantidad de dinero gastado por el usuario, el porcentaje de recomendaciÃ³n en base a reviews.recommend y cantidad de items.

def userdata(user_id): 
    ids = list(items_items[items_items["user_id"]==user_id]["item_id"])
    if len(ids) == 0:
        return print("No se encontro el ID")
    gastado = 0
    for id in ids:
        try:
            gastado += games[games["id"]==int(id)]["price"].values[0]
        except:
            continue
    gastado = round(gastado,2)
    porcentaje = (reviews_reviews[reviews_reviews["user_id"]==user_id]["recommend"].sum() / len(reviews_reviews[reviews_reviews["user_id"]==user_id])) *100
    print(f"El usuario {user_id} gasto ${gastado}")
    print(f"El porcentaje de recomendacion es {porcentaje}%")
    print(f"La cantidad de items que tiene es: {items[items['user_id']==user_id]['items_count'].values[0]}")
    return {"Gastado":gastado, "Porcentaje":porcentaje, "Cantidad de Items": {items[items['user_id']==user_id]['items_count'].values[0]}}
    
azar = reviews['user_id'].sample().values[0]
userdata(azar)


```
